```python
import yaml

# Chemin vers votre fichier data.yaml
file_path = '/kaggle/working/Swimming-Pools-2/data.yaml'

# Lire le fichier
with open(file_path, 'r') as file:
    data = yaml.safe_load(file)

# Supprimer la clé 'path'
if 'path' in data:
    del data['path']

# Modifier les chemins spécifiés
data['train'] = '/kaggle/working/Swimming-Pools-2/train/images'
data['val'] = '/kaggle/working/Swimming-Pools-2/valid/images'
data['test'] = '/kaggle/working/Swimming-Pools-2/test/images'

# Écrire les modifications dans le fichier
with open(file_path, 'w') as file:
    yaml.dump(data, file)

print("Modifications effectuées avec succès.")
```

```python
from ultralytics import YOLO
import shutil
import os

# Définir le chemin vers les fichiers de configuration du dataset
data_path = '/kaggle/working/Swimming-Pools-2/data.yaml'  # Assurez-vous que ce chemin est correct

# Initialiser le modèle YOLO (par exemple, YOLOv8s)
model = YOLO("yolov9t.yaml")  # Vous pouvez choisir un modèle pré-entraîné différent

# Entraîner le modèle
model.train(data=data_path, epochs=50, imgsz=640, batch=16, device=0)

# Chemin vers le répertoire de sauvegarde
save_dir = model.save_dir
print(f"Modèle sauvegardé dans : {save_dir}")
```
